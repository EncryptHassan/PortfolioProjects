# Exploratory Data Analysis (EDA)

## 1. What is the total sales across all walmart stores?
### **How it Helps:** Provides a bird's eye view of Walmart's overall product sales volume.
### **Business Impact:** Helps in determining high level sales performance and overall inventory demand.

```python
df['Quantity_Sold'].sum()
print("Total Quantity Sold across all Walmart stores:", df['Quantity_Sold'].sum())
```

### 2. What is the average unit price of all products sold?
### **How it Helps:** identifies product pricing trends across different categories.
### **Business Impact:** Assists in price optimization for different product lines to maximize profit margins.

```python
average_price = df.groupby('Product_Name')['Unit_Price'].mean()
```

```python
average_price
```

```python
 #
 plt.figure(figsize = (10,6))
 bars = average_price.plot(kind = 'bar')
 plt.title("Average Unit price of Products")
 plt.xlabel("Product Name")
 plt.ylabel("Average Unit price")
 plt.xticks(rotation = 45)
#  plt.show()

# add label details
 for index, value in enumerate(average_price):
  plt.text(index, value, f'{value:.2f}', ha  = 'center', va = 'bottom', fontsize = 10)
  plt.tight_layout()
plt.show()
```

### 3. Which stores are underperforming in terms of sales volume?
### **How it Helps:** Identifies stores that need additional inventory support or different marketing strategies.
### **Business Impact:** Allows the operations team to allocate inventory more strategically and suppot underperforming locations.

```python
sales_volume =  df.groupby('Store_Id')['Quantity_Sold'].sum().sort_values(ascending = True)
```

```python
print(sales_volume)
```

```python
 sales_threshold = sales_volume.median()
```

```python
sales_threshold
```

```python
# define underperforming stores as those below a threshold(e.g., below the median)
sales_threshold = sales_volume.median()
underperforming_stores = sales_volume[sales_volume < sales_threshold]

# Display underperforming stores
print("Underperforming Stores:", underperforming_stores)
```

### 4. What count of transactions involved promotions?
### **How it Helps:** Helps assess the reach and usage of promotional campaigns.
### **Business Impact:** Provides insight into whether promotions are driving enough sales to justify their costs.

```python
ax = sns.countplot(x = 'Promotion_Applied', data = df)
plt.title("Count of  Transactions with promotions")
plt.xlabel("Promotion Applied")
plt.ylabel("Count")
# plt.show()

# for data labels
for bar in ax.containers:
  ax.bar_label(bar)
  plt.tight_layout()
plt.show()
```

### 6. Which product categories see the largest increase in demand during holidays?
### **How it Helps:** Understands product demand variation during holidays.
### **Business Impact:** Enables better inventory planning for peak holidays seasons.

```python
df
```

```python
# Filter data for Holidays.
holiday_sales = df[df['Holiday_Indicator'] == True]

# Group by product_name an dcalculate total quantity sold
holiday_demand = holiday_sales.groupby('Product_Name')['Quantity_Sold'].sum()
holiday_demand
```

```python
# Plot the Bar graph.
plt.figure(figsize = (10,6))
ax = holiday_demand.plot(kind = 'bar')
plt.title("Product Categories with increased demand during Holidays")
plt.xlabel("Product Name")
plt.ylabel("Total Quatity Sold")
plt.xticks(rotation = 45)


# Add Data labels
for index, value in enumerate(holiday_demand):
  plt.text(index, value, f'{value:.2f}', ha = 'center', va = 'bottom', fontsize = 10)
  plt.tight_layout()
plt.show()
```

### 7. What is the correlation between weather conditions and sales performance?
### **How it Helps:** Reveals how external factors like weather influence customer purchasing behaviour.
### **Business Impact:** Assists in demand forecasting based on weather trends, leading to better inventory preparation.

```python
weather_impact = df.groupby('Weather_Conditions')['Revenue'].sum().sort_values()
weather_impact
```

```python
# PLot the Bar graph.
plt.figure(figsize = (10,6))
ax = weather_impact.plot(kind = 'bar')
plt.title("Impact of Weather Conditios on Revenue")
plt.xlabel("Weather Conditions")
plt.ylabel("Toatal Revenue")
plt.xticks(rotation = 45)

# Add labels.
# for bar in ax.containers:
#   ax.bar_label(bar)
#   plt.tight_layout()
# plt.show()

for index, value in enumerate(weather_impact):
    plt.text(index, value, f'{value:.2f}', ha = 'center', va = 'bottom', fontsize = 10)
    plt.tight_layout()
plt.show()
```

## 8. How does customer loyalty level affect purchasing patterns?

#### **How it helps:** Helps understand how loyalty programs impact repeat purchases.
#### **Business Impact:** Insights can be used to enhance customer retention strategies.

```python
# Perform groupby aggregation.
loyalty_analysis = df.groupby('Customer_Loyalty_Level').agg({
    'Quantity_Sold' : ['sum'],
    'Revenue' : ['sum', 'mean'],
    'Customer_Id' : 'nunique',
    'Transaction_Id' : 'count'
    }).reset_index()

# Sort the DataFrame by 'Revenue' (sum) in descending order
loyalty_analysis_sorted = loyalty_analysis.sort_values(by = ('Revenue', 'sum'), ascending = False)

# Display the sorted DataFrame
print(loyalty_analysis_sorted)
```

## 10. Which product has the best performance in terms of volume sold?
### **How it Helps:** Identifies top-performing products, revealing customer preferences an high demand items.

### **Business Impact** Supports targeted markrting efforts, inventory prioritization and strategic product placement to capatalize on best sellers an dimprove overall profitability.

```python
best_product = df.groupby('Product_Name')['Quantity_Sold'].sum().sort_values(ascending = False)
best_product
```
